```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

df = pd.read_csv('Civil_Engineering_Regression_Dataset.csv')

print("Column names in the dataset:")
print(df.columns)
df.columns = df.columns.str.strip()
if 'Building_Height' in df.columns:
    df = df.rename(columns={'Building_Height': 'Building Height'})

print("First few rows of the dataset:")
print(df.head())
X_simple = df[['Building Height']]
y = df['Construction_Cost']

X_train_simple, X_test_simple, y_train_simple, y_test_simple = train_test_split(X_simple, y, test_size=0.2, random_state=42)
model_simple = LinearRegression()
model_simple.fit(X_train_simple, y_train_simple)
y_pred_simple = model_simple.predict(X_test_simple)
r_squared_simple = r2_score(y_test_simple, y_pred_simple)

print(f"R-squared value for Simple Linear Regression: {r_squared_simple:.4f}")
X_multiple = df[['Building Height', 'Material_Quality_Index', 'Labor_Cost', 'Concrete_Strength', 'Foundation_Depth']]

X_train_multiple, X_test_multiple, y_train_multiple, y_test_multiple = train_test_split(X_multiple, y, test_size=0.2, random_state=42)
model_multiple = LinearRegression()
model_multiple.fit(X_train_multiple, y_train_multiple)

y_pred_multiple = model_multiple.predict(X_test_multiple)
r_squared_multiple = r2_score(y_test_multiple, y_pred_multiple)

print(f"R-squared value for Multiple Linear Regression: {r_squared_multiple:.4f}")
X_multiple_sm = sm.add_constant(X_multiple)
model_sm = sm.OLS(y, X_multiple_sm).fit()
adjusted_r_squared = model_sm.rsquared_adj

print(f"Adjusted R-squared value for Multiple Linear Regression: {adjusted_r_squared:.4f}")

vif_data = pd.DataFrame()
vif_data["Variable"] = X_multiple.columns
vif_data["VIF"] = [variance_inflation_factor(X_multiple.values, i) for i in range(X_multiple.shape[1])]

print("\nVariance Inflation Factor (VIF) for each variable:")
print(vif_data)
```
