```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Civil_Engineering_Regression_Dataset.csv')
print("First few rows of the dataset:")
print(df.head())


dependent_variable = 'Construction Cost'
independent_variables = df.columns[df.columns != dependent_variable].tolist()
print("\nIndependent Variables:", independent_variables)
print("Dependent Variable:", dependent_variable)

missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values[missing_values > 0])


for column in df.columns:
    if df[column].isnull().any():
        if df[column].dtype in ['int64', 'float64']:
            df[column].fillna(df[column].median(), inplace=True)
        else:
            df[column].fillna(df[column].mode()[0], inplace=True)

summary_statistics = df.describe()
print("\nSummary Statistics:")
print(summary_statistics)

correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
construction_cost_correlation = correlation_matrix['Construction_Cost']
print("\nCorrelation of Construction Cost with other factors:")
print(construction_cost_correlation)
```
